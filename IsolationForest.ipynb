```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Eliminar la columna 'quality' del anÃ¡lisis
X = df.drop(columns=['quality'])

#  Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#  Aplicar Isolation Forest
iso = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)  # 5% de outliers
df['anomaly'] = iso.fit_predict(X_scaled)

#  Mapear los resultados para facilitar la lectura
df['anomaly'] = df['anomaly'].map({1: 'Normal', -1: 'Outlier'})

# 5ï¸âƒ£ Ver cuÃ¡ntos outliers hay
print(df['anomaly'].value_counts())
```

```python
from sklearn.decomposition import PCA

# Reducimos a 2D para visualizar
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df['pca1'] = X_pca[:, 0]
df['pca2'] = X_pca[:, 1]

# GrÃ¡fico
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='pca1', y='pca2', hue='anomaly', palette={'Normal': 'blue', 'Outlier': 'red'})
plt.title("DetecciÃ³n de Outliers con Isolation Forest (PCA)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend(title="Tipo")
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Eliminar columnas solo si estÃ¡n presentes
cols_to_drop = ['quality', 'anomaly_score', 'anomaly_custom']
df_clean = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

# Asegurar que solo quedan columnas numÃ©ricas
X = df_clean.select_dtypes(include=['number'])

# Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Asegurar que X solo contenga columnas numÃ©ricas
X = X.select_dtypes(include=['number'])

# Escalar sin errores
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


#  Aplicamos Isolation Forest (sin usar contamination)
iso = IsolationForest(n_estimators=100, contamination='auto', random_state=42)
iso.fit(X_scaled)

#  Obtenemos los anomaly scores (cuanto mÃ¡s negativo, mÃ¡s outlier)
scores = iso.decision_function(X_scaled)  # Valores cercanos a 0 o negativos = mÃ¡s anormal
df['anomaly_score'] = scores

# Creamos un threshold (por ejemplo, 5% mÃ¡s bajo = outliers)
threshold = np.percentile(scores, 5)  # 5% peores (mÃ¡s negativos)
df['anomaly_custom'] = df['anomaly_score'].apply(lambda x: 'Outlier' if x < threshold else 'Normal')

#  Revisamos resultados
print(f"Umbral de outliers (5% mÃ¡s bajo): {threshold}")
print(df['anomaly_custom'].value_counts())
```

```python
import matplotlib.pyplot as plt

scores = iso.decision_function(X_scaled)  # Cuanto mÃ¡s negativo, mÃ¡s anÃ³malo

plt.figure(figsize=(8, 4))
plt.hist(scores, bins=50, color='skyblue')
plt.axvline(x=np.percentile(scores, 5), color='red', linestyle='--', label="umbral (5%)")
plt.title("DistribuciÃ³n de anomalÃ­a (Isolation Forest)")
plt.xlabel("Score (mÃ¡s negativo = mÃ¡s anÃ³malo)")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```
